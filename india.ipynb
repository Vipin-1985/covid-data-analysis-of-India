```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
df=pd.read_csv("/content/covid_19_india.csv",parse_dates=["Date"])
df.head()
```

```python
df.info()
```

```python
df.rename(columns={"Date":"date","State/UnionTerritory":"state_ut","ConfirmedIndianNational":"indian","ConfirmedForeignNational":"foregin","Cured":"cured","Deaths":"deaths","Confirmed":"confirmed","Time":"time"},inplace=True)
df.columns
```

```python
df.head()
```

```python
df["active"]=df["confirmed"]-df["deaths"]-df["cured"]
df.head()
```

```python
df.iloc[100:110,3:10]
```

```python
top=df[df["date"]==df["date"].max()]
top
```

```python
india=top.groupby("state_ut")["confirmed","deaths","cured"].sum().reset_index()
india
```

```python
total_cases=df.groupby("date")["date","active"].sum().reset_index()
total_cases
```

```python
plt.figure(figsize=(20,8))
plt.xticks(fontsize=12)
plt.yticks(fontsize=15)
plt.xlabel("Dates",fontsize=30)
plt.ylabel("Active cases",fontsize=30)
plt.title(" confirmed cases of india over time",fontsize=30)

ax=sns.lineplot(x=df.date,y=df.active,color="r")
#ax.set(xlabel="Dates",ylabel="Active cases")
```

```python
Active_cases=df.groupby("state_ut")["active"].sum().sort_values(ascending=False).reset_index()
Active_cases
```

```python
plt.figure(figsize=(25,15))
plt.xticks(rotation=90,fontsize=12)
plt.yticks(fontsize=15)
plt.xlabel("Active",fontsize=30)
plt.ylabel("State/UT",fontsize=30)
plt.title("top 20 countries with most active cases",fontsize=20)

ax=sns.barplot(x=Active_cases.active,y=Active_cases.state_ut)

```
